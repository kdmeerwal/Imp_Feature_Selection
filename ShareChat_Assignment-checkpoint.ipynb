```python
# 1.) Find the OrderID which has the maximum cost(value) , means highest billing order?
```

```python
Select Top 1 (
(Select OrderID, (p.Price/p.Unit)*OD.Quantity as price
 From OrderDetails OD, Products P
 Left Join OrderDetails OD On OD.ProductID = P.ProductID
 Group by OrderID
 Order by price DESC)a)
```

```python
# 2.) Which country generated the maximum revenue?
```

```python
Select * into temp1 From (
(Select OrderID, (p.Price/p.Unit)*OD.Quantity as Revenue
 From OrderDetails OD, Products P
 Left Join OrderDetails OD On OD.ProductID = P.ProductID
 Group by OrderID)a)

Select * into temp2 From (
(Select O.OrderID, C.Country
 From Orders O, Customers C
 Left Join Orders O On O.CustomerID = C.CustomerID
 Group by O.OrderID)b)

Select Top 1 (
(Select sum(temp1.Revenue) as Max_Revenue , temp2.Country 
 From temp1 , temp2 
 Left Join temp1 On temp1.OrderID = temp2.OrderID
 Group by temp2.Country, 
 Order by Max_Revenue DESC)c)
```

```python
# 3.) Among the people who bought product X(productid=x) on date 2019-10-01, find how many did buy the product x on the following day?
```

```python
Select * into temp From (
Select a.CustomerID as Cust_ID, a.OrderDate as Date
From Orders a 
Left Join OrderDetails as b 
On a.OrderID = b.OrderID
where a.OrderDate Between '2019-10-01' And '2019-10-02' And b.ProductID = 'x'
)

Select * into temp1 From (
Select Cust_ID From temp
where Date = '2019-10-01'
Group By Cust_ID
)

Select * into temp2 From (
Select Cust_ID From temp
where Date = '2019-10-02'
Group By Cust_ID
)

Select Distinct Count(ID) From ( 
Select g.Cust_ID as ID
From temp2 as g
Inner Join temp1 as f
on g.Cust_ID = f.Cust_ID
)
```

```python
# Access https://jsonplaceholder.typicode.com/albums/1/photos through https://jsonplaceholder.typicode.com/albums/100/photos 
# which collectively contain 5000 photos in JSON format using an HTTP GET request. 
# Group all the photos into categories based on the first letter of their title. 
# Return the number of photos with each starting letter, and plot as bar-chart.
```

```python
import urllib.request
import json
import matplotlib.pyplot as plt
```

```python
init_letter = {}

for x in range(1,101):
    webUrl = urllib.request.urlopen('https://jsonplaceholder.typicode.com/albums/' + str(x) +  '/photos')
    encoding = webUrl.info().get_content_charset('utf-8')
    data = webUrl.read()
    data = json.loads(data.decode(encoding))
    for i in range(0,len(data)):
        print((data[i]['title'][0]))
        letter = data[i]['title'][0]
        if ( letter not in init_letter.keys() ):
            init_letter[letter] = 1
        else :
            init_letter[letter] += 1
```

```python
print(init_letter)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
keys = init_letter.keys()
values = init_letter.values()
ax.bar(keys, values)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)
```

```python
# Create linear regression object
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
regr = LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()

from sklearn.metrics import mean_squared_error, r2_score
```

```python
rfr.fit(X_train, y_train)
```

```python
# Make predictions using the testing set
y_pred = rfr.predict(X_test)

# The coefficients
print('Coefficients: \n', rfr.feature_importances_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_pred))
```

```python
df = pd.read_csv('city_weather.csv')
df = df.drop(['pressure', 'sea_level', 'country', 'lat', 'lon'], axis = 1)

y = df['temp_min']
x = df.drop(['temp_min'], axis = 1)
```

```python
model = LinearRegression(normalize=True)
model = model.fit(x,y)
print(model.score(x,y))
print(model.coef_)
```

```python
## this above models tells me wind_speed is most important factor with positive correlation
## wind_speed > rain > humidity > grnd_level > clouds
```

```python
### find change in r2 when added last

def getAdjustedR2(r2, n , p):
    return 1 - (1 - r2)*(n-1)/(n-p-1)
```

```python
r2_change = {}

for col in x.columns:
    x_without = x.drop([col], axis=1)
    model1 = LinearRegression(normalize=True).fit(x_without, y)
    model2 = LinearRegression(normalize=True).fit(x, y)
    r2_change[col] = getAdjustedR2(model2.score(x,y), x.size, len(x.columns)) - \
        getAdjustedR2(model1.score(x_without, y), x_without.size, len(x_without.columns))

print(r2_change)
```
